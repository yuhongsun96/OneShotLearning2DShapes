# One Shot Learning of 2D Shapes

## Introduction

Modern computer vision is centered around deep learning which requires extensive datasets. Here we showcase a different approach for identifying 2D shapes in a translation, rotation, and scale invariant fashion while only using a single example per class for training.

```python
from matplotlib import pyplot as plt
from matplotlib import image as mpimg 
from matplotlib.patches import Polygon
import cv2
import numpy as np
import random
import math
import khiva as kv
from khiva.distances import sbd
```

```python
# Set the below to KHIVABackend.KHIVA_BACKEND_CUDA if GPU is available
kv.set_backend(kv.KHIVABackend.KHIVA_BACKEND_CPU)
```

```python
plt.rcParams['figure.figsize'] = [10, 8]    # Set the size for the visualizations in this notebook
pixel_thresh = 127                          # Grayscale pixel value cutoff
pixel_max = 255                             # Maximum value for pixel
degrees = 360                               # Time slicing images by 1 degree per timestamp
dpi = 1000                                  # Resolution of saved png for training/testing
```

## Generate the training set

We generate a training set with 1 example per class. Each shape is generated with a different size, rotation and center location then saved as an image under train_data. The randomly generated shapes can be also replaced with custom (user provided) grayscale images. The image generation code is simply provided for convenience.

```python
# Example training data set with format: (number of sides, file name, class name)
train_shapes = [
    (3, "train_data/triangle.png", "Triangle"),
    (4, "train_data/square.png", "Square"),
    (5, "train_data/pentagon.png", "Pentagon")
]
```

```python
# Generates a 2d polygon with a random size, rotation, and location within a unit square
# Takes a specified number of sides and returns the coordinates of the vertices
def gen_shape(num_sides):
    vertices = []
    angles = degrees / num_sides
    
    # Set a center, starting angle and radial length for the polygon
    starting_angle = np.random.rand() * degrees
    max_length = 0
    while max_length < 0.1:
        center = np.random.rand(2)
        max_length = min(center[0] - 0, 1 - center[0], center[1] - 0, 1 - center[1])
    radial_length = random.uniform(0.1, max_length)
    
    # Generate vertices
    for point in range(num_sides):
        angle = starting_angle + point * angles
        x = center[0] + np.sin(np.deg2rad(angle)) * radial_length
        y = center[1] + np.cos(np.deg2rad(angle)) * radial_length
        vertices.append([x, y])
    
    return vertices
```

```python
# Save the generated shape as png, the starting point of the algorithm is an image
def save_shape(vertices, name):
    # Make sure the shapes are fully shaded in
    pol = Polygon(vertices, True, edgecolor='black', facecolor='black')
    fig,ax = plt.subplots()
    # Ensure aspect ratio
    ax.set_aspect('equal', 'box')
    plt.axis('off')
    currentAxis = plt.gca()
    currentAxis.add_patch(pol)
    plt.savefig(name, dpi=dpi)
```

```python
%%capture
# Generate the shapes and save them to the training set
for shape in train_shapes:
    save_shape(gen_shape(shape[0]), shape[1])
#If desired, custom shapes can be added to train_shapes below
```

### The graphic below shows an example of one of the generated shapes

```python
# We'll use this index throughout the notebook as the training example to visualize
example = random.choice(range(len(train_shapes))) 
print(plt.imshow(mpimg.imread(train_shapes[example][1])))
```

## Discovering the shape within the image

Now that the training dataset has been generated, we begin by centering the shape within the image using openCV.

```python
# Given an image file, returns:
#     the matrix of the image
#     the coodinates (x, y) of the centroid of the shape
#     the radius of the minimal encompassing circle (only used for visualization)
def get_centroid(file):
    img = cv2.imread(file, cv2.IMREAD_GRAYSCALE)
    ret, thresh = cv2.threshold(img, pixel_thresh, pixel_max, cv2.THRESH_BINARY_INV)
    contours, hierarchy = cv2.findContours(thresh, 1, 2)
    cnt = contours[0]
    (x, y), radius = cv2.minEnclosingCircle(cnt)
    center = (int(x), int(y))
    radius = int(radius)
    return img, center, radius
```

```python
# Read in all the training images
images, centroids, radiuses = [], [], []
for shape in train_shapes:
    img, cent, rad = get_centroid(shape[1])
    images.append(img)
    centroids.append(cent)
    radiuses.append(rad)
```

### The graphic below shows the example with its minimal encompassing circle

```python
print(plt.imshow(cv2.circle(images[example], centroids[example], radiuses[example], (0, pixel_max, 0), 2), "gray"))
```

## Generate timeseries from shape

Timeseries data is generated by scanning outward from the centroid of the shape (found in the previous step) at different angles. Each angle/scan represents a single timestep. In this example we will scan once at each degree resulting in a 360 step timeseries. The value of a particular timestep is the distance from the centroid to the edge of the shape at the corresponding angle.

```python
def gen_timeseries(img, cent):
    times = []
    # Iterates every angle, may be parallelizable in future
    for deg in range(degrees):
        dx = np.sin(np.deg2rad(deg))
        dy = np.cos(np.deg2rad(deg))
        step = 0
        while True:
            x_val = int(cent[0] + round(step * dx))
            y_val = int(cent[1] + round(step * dy))
            # The edge case below may happen if user provides invalid training/test image
            if y_val < 0 or y_val >= img.shape[0] or x_val < 0 or x_val >= img.shape[1]:
                print("Shape does not fit completely within image")
                return None
            # If pixel value changed in the last timestep, edge is assumed to have been reached
            if img[y_val][x_val] > pixel_thresh:
                times.append(step)
                break
            step += 1

    return times
```

### The graphic below shows the timeseries generated from the training examples
The timeseries from all of the training images are shown in the plot below. The example from the previous graphics is shown in bold.

```python
# Plot the lines for the timeseries generated from the training shapes
train_timeseries = []
for ind in range(len(train_shapes)):
    train_timeseries.append(gen_timeseries(images[ind], centroids[ind]))
    if ind == example:
        plt.plot(train_timeseries[ind], linewidth=7.0)
    else:
        plt.plot(train_timeseries[ind])
```

## Generate the test example
The test shape is generated with a random number of sides, location, size, and rotation and saved as an image. This test shape can also be provided as an image intead of being generated.

```python
%%capture
test_image_name = "test_data/unknown.png"
# Select a number of sides that exists in the training set
range_sides = set([exp[0] for exp in train_shapes])
test_sides = random.choice(list(range_sides))
# Save the generated test shape
save_shape(gen_shape(test_sides), test_image_name)
```

### The graphic below shows the generated test image

```python
print(plt.imshow(mpimg.imread(test_image_name)))
```

## Generate timeseries for the test example
We use the same methodology as the training steps for converting an image into a timeseries.

```python
test_img, test_centroid, _ = get_centroid(test_image_name)
test_timeseries = gen_timeseries(test_img, test_centroid)
```

### The graphic below shows the timeseries created from the test image
You may note that the periods and amplitude of this timeseries don't necessarily match any of the training examples. In fact it could be an arbitrarily complex sequence if a corresponding complex shape is provided. As long as the general shape of the timeseries and its motifs can be matched to training sequences, the image can be classified.

```python
print(plt.plot(test_timeseries))
```

## Matrix profile and comparing timeseries
Now that we have the timeseries of the training and test shapes, we can use SBD to compare the timeseries. SBD is a specialized timeseries distance metric which compares based on the shape of the timeseries so that we can match across varying periods and scales. We use the Khiva implementation here.

```python
combined_timeseries = [test_timeseries] + train_timeseries
# khiva sbd gives back a triangular matrix with distances between timeseries as elements of the matrix
result_matrix = sbd(kv.Array(combined_timeseries)).to_numpy()
```

### The matrix below is the distance matrix between the timeseries

```python
print(result_matrix)
```

## Classification
For this demo, we choose to match the test shape to the closest training shape. Using the same methodology, we can also set a similarity distance threshold and see which, if any, of the training shapes the test shape is sufficiently similar to.

```python
# The distances of the test timeseries to the rest is represented by the first column of the matrix
# We match it to the closest fit from the training set and select the class label provided
classification = train_shapes[np.argmin(result_matrix[1:, 0])][2]
```

```python
print("The final classification of the generated shape with " + str(test_sides) +" sides is:\n" + classification)
```
