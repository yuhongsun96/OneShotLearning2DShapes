```python
from matplotlib import pyplot as plt
from matplotlib.patches import Polygon
import cv2
import numpy as np
import random
import math
import khiva as kv
```

```python
kv.set_backend(kv.KHIVABackend.KHIVA_BACKEND_CPU)
```

```python
pixel_thresh = 127
pixel_max = 255
degrees = 360
dpi = 1000
kshape_iters = 10
```

```python
train_shapes = [
    (3, "train_data/triangle1.png", "Triangle"),
    (3, "train_data/triangle2.png", "Triangle"),
    (4, "train_data/square1.png", "Square"),
    (4, "train_data/square2.png", "Square"),
    (5, "train_data/pentagon1.png", "Pentagon"),
    (5, "train_data/pentagon2.png", "Pentagon")
]
```

```python
labels = [exp[2] for exp in train_shapes]
num_classes = len(set(labels))
```

```python
def gen_shape(num_sides):
    vertices = []
    angles = degrees / num_sides
    
    # Set a center, starting angle and radial length for the polygon
    starting_angle = np.random.rand() * degrees
    max_length = 0
    while max_length < 0.1:
        center = np.random.rand(2)
        max_length = min(center[0] - 0, 1 - center[0], center[1] - 0, 1 - center[1])
    radial_length = random.uniform(0.1, max_length)
    
    # Generate vertices
    for point in range(num_sides):
        angle = starting_angle + point * angles
        x = center[0] + np.sin(np.deg2rad(angle)) * radial_length
        y = center[1] + np.cos(np.deg2rad(angle)) * radial_length
        vertices.append([x, y])
    
    return vertices
```

```python
def save_shape(vertices, name):
    pol = Polygon(vertices, True, edgecolor='black', facecolor='black')
    fig,ax = plt.subplots()
    ax.set_aspect('equal', 'box')
    plt.axis('off')
    currentAxis = plt.gca()
    currentAxis.add_patch(pol)
    plt.savefig(name, dpi=dpi)
```

```python
%%capture
for shape in train_shapes:
    save_shape(gen_shape(shape[0]), shape[1])
```

```python
def get_centroid(file):
    img = cv2.imread(file, cv2.IMREAD_GRAYSCALE)
    ret, thresh = cv2.threshold(img, pixel_thresh, pixel_max, cv2.THRESH_BINARY_INV)
    contours, hierarchy = cv2.findContours(thresh, 1, 2)
    cnt = contours[0]
    (x, y), radius = cv2.minEnclosingCircle(cnt)
    center = (int(x), int(y))
    radius = int(radius)
    return img, center, radius
```

```python
images, centroids, radiuses = [], [], []
for shape in train_shapes:
    img, cent, rad = get_centroid(shape[1])
    images.append(img)
    centroids.append(cent)
    radiuses.append(rad)
```

```python
example = random.choice(range(len(train_shapes)))
plt.imshow(cv2.circle(images[example], centroids[example], radiuses[example], (0, pixel_max, 0), 2), "gray")
```

```python
#non-optimized cpu implementation
def gen_timeseries(img, cent):
    times = []
    for deg in range(degrees):
        dx = np.sin(np.deg2rad(deg))
        dy = np.cos(np.deg2rad(deg))
        step = 0
        while True:
            x_val = int(cent[0] + round(step * dx))
            y_val = int(cent[1] + round(step * dy))
            if y_val < 0 or y_val >= img.shape[0] or x_val < 0 or x_val >= img.shape[1]:
                print("Shape does not fit completely within image")
                return None
            if img[y_val][x_val] > pixel_thresh:
                times.append(step)
                break
            step += 1

    return times
```

```python
training_timeseries = []
for ind in range(len(train_shapes)):
    training_timeseries.append(gen_timeseries(images[ind], centroids[ind]))
    plt.plot(training_timeseries[ind])
```

```python
(centroids, indices) = kv.k_shape(kv.Array(training_timeseries), num_classes, max_iterations=kshape_iters)
```

```python
data = np.asarray(training_timeseries)
np_centroids = centroids.to_numpy()
ind = indices.to_numpy()
first_cluster = data[ind == 0]
plt.plot(first_cluster.T)
```
